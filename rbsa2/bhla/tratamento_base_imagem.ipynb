```python
#import
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

**Colocando o tamanho da imagem no data set**

```python
# reading multimodal B-T4SA dataset
path = "D:/SiDi/Project/Modulo II/dataset/output/dataset_sidi.csv"
df = pd.read_csv(path, sep='\t')
df.head()
```

```python
#adding height and width to multimodal B-T4SA dataset
path_image = "D:/SiDi/Project/Modulo II/dataset/b-t4sa_imgs/"
image_width = []
image_height = []

for i in df["image_path"]:

    img = cv2.imread(path_image+i)
    aux = img.shape
    image_height.append(aux[0])
    image_width.append(aux[1])
    
df['image_height'] = image_height
df['image_width'] = image_width
df.head()
```

```python
#exporting to multimodal B-T4SA dataset with adding features (height and width)
df.to_csv("D:/SiDi/Project/Modulo II/dataset/output/dataset_sidi_v1.2.csv", sep='\t', encoding='utf-8')
```
